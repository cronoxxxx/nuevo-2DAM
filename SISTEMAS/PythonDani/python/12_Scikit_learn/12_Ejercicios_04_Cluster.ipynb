Tomamos la siguiente distribuci√≥n de puntos:

```python
def load2_points(N):
    P = np.zeros([N,2])
    for j in range(N):  
        a=2*math.pi*random.random()
        R=random.choice([40,42,44,5])
        R=R+20*random.random()
        P[j,0]=(R*math.sin(a))
        P[j,1]=(R*math.cos(a))
    return P


N=1000
P=load2_points(N)
plt.figure(figsize=(5, 5))
plt.scatter(P[:, 0], P[:, 1],s=8, alpha=0.4)
```
