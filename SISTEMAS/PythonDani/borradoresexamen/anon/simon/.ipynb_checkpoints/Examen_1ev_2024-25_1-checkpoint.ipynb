Dibuja las siguientes funciones:
$$ y = A*x $$

* (0.5 puntos) Los colores de las gráficas son verde, rojo y azul, la linea de 1 y el grid que aparezca por defecto

* (0.5 puntos) Marca la solución con una $\star$ y pon las etiquetas en cada eje


<img src="matplotlib.png" ALIGN="left" width=200 height=200)>


````python
import numpy as np
import matplotlib.pyplot as plt
#np.arange(0, 1.5, 0.05) genera un arreglo de valores que comienza en 0, termina en 1.5 (exclusivo) y se incrementa en pasos de 0.05. 
x = np.arange(0,1.5, 0.05)
y = A * x
y2 = A * (x*x)
#cambiando la fuente y poniendo el titulo
font = {'family':'arial','color':'red','size':16}
plt.title("Ignacio Llorente", loc = 'center', fontdict = font)

#Poniendo x e y
plt.plot(x,y,linewidth = '1',color='g',label='y = 17*x')
plt.plot(x,y2,linewidth = '1',color='r',label='y = 17*$x^2$')
plt.grid()

#Poniendo la solución
plt.plot([1,1],[0,17],color='b',label='solucion')
plt.legend()

#Poniendo la estrellita en la solución
plt.plot([1,1],[17,17],color='b',marker = '*')

plt.xlabel("x")
plt.ylabel('y')

````

* Crea una clase llamada Repartidor.
* El constructor inicializará los atributos **(nombre, zona_entrega, pedidos_entregados)** cuando se cree una instancia de la clase.
* Crea un método **entregar_pedido** que registra la entrega de un pedido y muestra un mensaje.
* Crea un método **informe_entregas** que imprime un informe con el número total de pedidos entregados por el repartidor.

Para probar la clase utiliza el sigiente código:

```

repartidor1.entregar_pedido()

```

Deberías obtener:
```

Carlos ha entregado un pedido en la zona Centro.
Carlos ha entregado un pedido en la zona Centro.
Carlos ha entregado 3 pedidos en total.
```

````python
class Repartidor:
    def __init__(self,nombre, zona_entrega, pedidos_entregados = 0):
        self.__nombre = nombre; 
        self.__zona_entrega = zona_entrega;
        self.__pedidos_entregados = pedidos_entregados
        
    def entregar_pedido(self):
        self.__pedidos_entregados = self.__pedidos_entregados + 1
        realizado = self.__nombre + " ha entregado un pedido en la zona " +  self.__zona_entrega + "."
        print(realizado)

    def informe_entregas(self):
        print(self.__nombre + " ha entregado " + str(self.__pedidos_entregados) + " pedidos en total.")
    

#Descomenta el siguiente código para probar la clase:

repartidor1 = Repartidor("Carlos", "Centro")

repartidor1.entregar_pedido()
repartidor1.entregar_pedido()
repartidor1.entregar_pedido()

repartidor1.informe_entregas()
````
